```python
df_kmean=df.copy()
df_kmean
```

```python
df_kmean = df_kmean.drop(['Row ID','Order ID','Customer ID','Country','Product ID'],axis=1)
```

```python
df_k = df_kmean[['Ship Mode','Quantity','Sales','Profit','Sub-Category','Category','Region']].copy()
```

```python
sub_Cat = df_k['Sub-Category'].astype('category')
sub_Cat_dict = dict(enumerate(sub_Cat.cat.categories))
print(sub_Cat_dict)
category_cat = df_k['Category'].astype('category')
cat_dict = dict(enumerate(category_cat.cat.categories))
print(cat_dict)
region_cat = df_k['Region'].astype('category')
region_dict = dict(enumerate(region_cat.cat.categories))
print(region_dict)
ship_cat = df_k['Ship Mode'].astype('category')
ship_dict = dict(enumerate(ship_cat.cat.categories))
print(ship_dict)
```

```python
# first we reverse our dictionaries to apply them on the dataframe 

sub_map = {v: k for k, v in sub_Cat_dict.items()}
cat_map = {v: k for k, v in cat_dict.items()}
region_map = {v: k for k, v in region_dict.items()}
ship_map = {v: k for k, v in ship_dict.items()}
df_k = df_k.replace({'Ship Mode':ship_map,'Region':region_map,'Category':cat_map,'Sub-Category':sub_map})
df_k
```

```python
df_k = df_k.dropna()
df_k
```

```python
from dash.dcc import Graph
from dash.html import Center
from dash.html.Figure import Figure

# scatter to show the relation betwwn sales and profits 
# Sales & Profits by States 
fig=px.scatter(df,x="Sales",y="Profit",color="City",title = ' Sales & Profits by States',
              size="Quantity",size_max=60,hover_name="Country",log_x=True,
              width=900,height=450,animation_frame="Segment")

# strip sales and profit for each region 
# Sales & Profit Distribution<br><i><i>Grouped By Region
fig1=px.strip(df,x="Sales",y="Profit", color="City", facet_col="Region" , title = "Sales & Profit Distribution By Region ")

# pie chart for percentage for Sub-Category
fig2=px.pie(df['Profit'],names=df['Sub-Category'],width=650,height=350,title='Sales analysis for each Sub_Category.')

# bar to show Sub-Category and Category
fig3=px.bar(df, x='Sub-Category', y='Sales')

# line to show the relation between sales and profit 
fig4=px.line(data,x ='month',y= ["Profit","Sales"] , title='Sales /profit monthly. ') 

app = JupyterDash(external_stylesheets=[dbc.themes.JOURNAL])

app.layout = dbc.Container([
                             dbc.Row ( 
                                [dbc.Col([
                              html.H6("City", style={'font-size': '20px','color': 'white', 'font-family':'Sans-serif'}),
                              html.H1(city, id='city', style={'font-size': '30px','color': 'white'})] , 
                                      width=4, style={"height": "180px", 'font-family': 'Cursive', "background-color": "#37597d",
                                      'text-align':'center', 'justify':'center', 'align':'center', 'padding-top': '40px'}),

                         
                        dbc.Col([
                            html.H6("Profit", style={'font-size': '20px','color': 'white', 'font-family':'Sans-serif'}),
                            html.H1(Profit, id='profit', style={'font-size': '30px','color': 'white'})],
                            width=4, style={"height": "180px", 'font-family': 'Cursive', "background-color": "#37597d", 
                              'text-align':'center', 'justify':'center', 'align':'center', 'padding-top': '40px'}),

                        
                        dbc.Col([
                            html.H6("Year", style={'font-size': '20px','color': 'white', 'font-family':'Sans-serif'}),
                            html.H1(year, id='year', style={'font-size': '30px','color': 'white'})],
                            width=4, style={"height": "180px", 'font-family': 'Cursive', "background-color": "#37597d", 
                                           'text-align':'center', 'justify':'center', 'align':'center', 
                                            'padding-top': '40px'}),

                    ] ),
                           
                            #__________________#    
                             dbc.Row(
                                  [
                                  dbc.Col(html.Div([dcc.Graph(figure = fig)]),width=5,style={"height": "100%"}),
                                  
                                  dbc.Col(html.Div([dcc.Graph(figure = fig4)])),
                                  ]),
                            #____________________#
                            dbc.Row(html.Div([dcc.Graph(figure = fig1)])),
                            #____________________#
                            dbc.Row(
                                 [
                                  dbc.Col(html.Div(dcc.Graph(figure = fig2))),
                                                  
                                  dbc.Col(html.Div(dcc.Graph(figure = fig3))),
        
                                 ]),
                            #____________________#
                             dbc.Row(
            [
                dbc.Col(controls, md=4),
                dbc.Col(dcc.Graph(id="cluster-graph"), md=8),
            ],
            align="center",
        )
    ])
           
       

@app.callback(
    Output("cluster-graph", "figure"),
    [
        Input("x-variable", "value"),
        Input("y-variable", "value"),
        Input("cluster-count", "value"),
    ],
)
def make_graph(x, y, n_clusters):
    # minimal input validation, make sure there's at least one cluster
    km = KMeans(n_clusters=max(n_clusters, 1))
    df = df_k
    km.fit(df)
    df["cluster"] = km.labels_

    centers = km.cluster_centers_

    data = [
        go.Scatter(
            x=df.loc[df.cluster == c, x],
            y=df.loc[df.cluster == c, y],
            mode="markers",
            marker={"size": 8},
            name="Cluster {}".format(c),
        )
        for c in range(n_clusters)
    ]

    data.append(
        go.Scatter(
            x=centers[:, 0],
            y=centers[:, 1],
            mode="markers",
            marker={"color": "#000", "size": 12, "symbol": "diamond"},
            name="Cluster centers",
        )
    )

    layout = {"xaxis": {"title": x}, "yaxis": {"title": y}}

    return go.Figure(data=data, layout=layout)


# make sure that x and y values can't be the same variable
def filter_options(v):
    """Disable option v"""
    return [
        {"label": col, "value": col, "disabled": col == v}
        for col in columns
    ]


# functionality is the same for both dropdowns, so we reuse filter_options
app.callback(Output("x-variable", "options"), [Input("y-variable", "value")])(
    filter_options
)
app.callback(Output("y-variable", "options"), [Input("x-variable", "value")])(
    filter_options
)

                    
app.run_server()
```
